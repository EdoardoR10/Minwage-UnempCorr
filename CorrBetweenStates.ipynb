```python
df.drop(columns={'Footnote','High.Value','Low.Value','Unnamed: 0','Table_Data'},inplace=True)
```

```python
df.head()
df.set_index('Year',inplace=True)
df.sort_index(inplace=True)
```

```python
df.head()
```

```python
df=df[df.index>=1990]
```

```python
import numpy as np
```

```python
dframe=pd.DataFrame()
for name,group in df.groupby('State'):
    if dframe.empty:
        dframe=group[['Low.2018']].rename(columns={'Low.2018':name})
    else:
        dframe[f'{name}']=group['Low.2018']
for columns in dframe:
    if dframe[columns].sum()==0:
        dframe.drop(columns,axis=1,inplace=True)
dframe.head()
```

```python
def get_val(year,state):
    try:
        return dframe.loc[year][state]
    except:
        return np.NaN
get_val(2012,'Arizona')
```

```python
%%time
df2['min_wage']=list(map(get_val,df2.index,df2['State']))
```
